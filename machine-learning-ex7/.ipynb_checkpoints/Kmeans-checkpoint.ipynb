### 1.1 Implementing K means

#### 1.1.1 Finding closest centroids

```python
import scipy.io
import numpy as np
import math
from scipy.spatial import distance
import matplotlib.pyplot as plt
import imageio.v2 as imageio
from platform import python_version
print(python_version())
```

```python
# Load example dataset
data = scipy.io.loadmat('ex7data2.mat')
type(data)
data.keys()
```

```python
X = data['X']
X.shape
```

```python
# Select an initial set of centroids
# (3,3) a point in 2d because our input is 2d.
K = 3
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])
```

```python
def find_closest_centroids(X, centroids):
    
    K = centroids.shape[0] # number of centroids
    idx = np.zeros([X.shape[0], 1])
    
    for i in range (X.shape[0]):
        meanArray = np.mean(np.square((centroids - X[i,])), axis=1)
        minval_index = np.argmin(meanArray)
        idx[i] = minval_index
        
    return idx
```

```python
idx = find_closest_centroids(X, initial_centroids)
```

```python
print('Closest centroids for the first 3 examples:')
print(idx[0:3])

#in octave its 1,3,2 so in Python its 0, 2, 1
print('Closest centroids should be 0, 2, 1 respectively')
```

#### 1.1.2 Computing centroid means

```python
def computeCentroids(X, idx, K):
    
    m, n = X.shape[0], X.shape[1]
    centroids = np.zeros([K, n])
    
    #find the positions/indices where centroid is mapped as 1st, 2nd , -- K, and then find the mean of all those points.
    #This the new centroid.
    
    for i in range (K) :
        pos=np.where(idx[:,0] == i)
        X_mapd_to_centd_i = X[pos,]
        X_mapd_mean = np.mean(X_mapd_to_centd_i, axis=1)
        centroids[i,] = X_mapd_mean
    return centroids
        
```

```python
centroids = computeCentroids(X, idx, K)
```

```python
print('Centroids computed after initial finding of closest centroids:')
print(centroids)
print('\nExpected centroids:');
print('[ 2.428301 3.157924 ]');
print('[ 5.813503 2.633656 ]');
print('[ 7.119387 3.616684 ]');
```

### 1.2 *K*-means on example dataset

```python
#Load an example dataset
# Loaded initially scipy.io.loadmat('ex7data2.mat')

#Settings for running K-Means
K = 3
max_iters = 10

# For consistency, here we set centroids to specific values
# but in practice you want to generate them automatically, such as by
# settings them to be random examples (as can be seen in
# kMeansInitCentroids).
initial_centroids = np.array([[3,3], [6,2], [8,5]])
```

```python
def draw_line(p1, p2):
    plt.plot(np.array([p1[0], p2[0]]), np.array([p1[1], p2[1]]), c='black', linewidth=1)
```

```python
def plotProgresskMeans(X, centroids, previous, idx, K, i):
    
    
    plt.scatter(X[:, 0], X[:, 1], marker='o', c=idx, cmap='rainbow')
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', c='black', s=25)
   
    for j in range(centroids.shape[0]):
        draw_line(centroids[j], previous[j])
  
    
```

```python
def runkMeans(X, initial_centroids, max_iters, plot_progress=False):
    
    #Set initial variables
    m, n = X.shape[0], X.shape[1]
    centroids = initial_centroids
    previous_centroids = centroids
    K = centroids.shape[0]
    idx = np.zeros([m, 1])
       
    for i in range (max_iters) :
        
        print('K-Means iteration' , i+1 , max_iters)
        idx = find_closest_centroids(X, centroids)
        
        if plot_progress:
            plotProgresskMeans(X, centroids, previous_centroids, idx, K, i)
            previous_centroids = centroids
            #input('Press ENTER to continue')
         #Compute Centroids again.
        centroids = computeCentroids(X, idx, K)
        plotProgresskMeans(X, centroids, previous_centroids, idx, K, i)
    
    return centroids, idx
```

```python
centroids, idx = runkMeans(X, initial_centroids, max_iters, plot_progress=True)
```

### 1.3 Random initialization


```python
def kMeansInitCentroids(X, K):
    
    centroids = np.zeros([K, X.shape[1]])
                              
    #Randomly reorder the indices of examples
    randidx = np.random.permutation(X.shape[0])
                              
     # Take the first K examples as centroids
    centroids = X[randidx[:K], :]
                              
    return centroids
    
```

```python
initial_centroids = kMeansInitCentroids(X, K)
centroids, idx = runkMeans(X, initial_centroids, max_iters, plot_progress=True)
```

### 1.4 Image compression with K-means

#### 1.4.1 *K*-means on pixels

```python
#imageio.imread(uri, format=None, **kwargs)
#Reads an image from the specified file. Returns a numpy array, which comes with a dict of meta data at its meta attribute.
image = imageio.imread('bird_small.png')
plt.imshow(image)
plt.show()
```

```python
type(image)
```

```python
#size 128x128, where each pixel has three values (0-255)
# for it's red, green and blue intensities
image.shape
```

```python
# Divide by 255 so that all values are in the range 0 - 1
A = image / 255 
```

```python
# Reshape the image into an Nx3 matrix where N = number of pixels.
# Each row will contain the Red, Green and Blue pixel values
# This gives us our dataset matrix X that we will use K-Means on.
#my note - 1st column - 1st layer which is Red, 2nd for blue, 3rd column for green,
X = np.reshape(A, [A.shape[0] * A.shape[1], 3])
```

```python
# Run k-means
K = 16
max_iters = 10

initial_centroids = kMeansInitCentroids(X, K)
centroids, idx = runkMeans(X, initial_centroids, max_iters)
```

```python
idx
```

```python
idx = idx.astype(int)
idx
```

```python
X_recovered = centroids[idx]
X_recovered
```

```python
#Reshape back to 128 x 128
X_recovered = np.reshape(X_recovered, image.shape)
X_recovered
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.title('Original image in 24-bit colour')
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(X_recovered)
plt.title('Compressed image in 4-bit colour')
plt.show()
```
